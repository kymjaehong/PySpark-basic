````python
from pyspark import SparkContext
````

````python
sc= SparkContext('local')
````

````python
rdd1= sc.textFile('data/bb.txt')
rdd1.collect()
````

## Transformations

````python
rdd2= rdd1.flatMap(lambda v: v.split(','))
rdd2.collect()
````

````python
rdd3= rdd2.map(lambda v: int(v))
rdd3.collect()
````

#### .distinct( )
- 중복 삭제

````python
rdd4= rdd3.distinct()
rdd4.collect()
````

### transformations 함수 + 함수

````python
myr= rdd1.flatMap(lambda v: v.split(',')).map(lambda v: int(v)).distinct()
myr.collect()
````

## Actions
```
 first, sum, max, mean, stdev...
 ```

````python
myr.first()
````

````python
myr.sum()
````

````python
myr.mean()
````

````python
myr.stdev()
````

#### .histogram([a,b,c] )
- a<= data<= b
- b<= data<= c
- 참조 변수의 값들이 각 범위 당 몇 개씩 있는지 반환
- 하나의 값-> 몇 등분 할 지-> 자동 분리-> 비교-> 출력

````python
myr.histogram([1,50,100])
````

````python
myr.stats()
````

#### .reduce(func, func)
- 누적 값
```
myr= [10, 20, 30, 40, 50, 60]
10, 20, 
10+20, 30,
10+20+30, 40,
10+20+30+40, 50,
10+20+30+40+50, 60,
10+20+30+40+50+60
```


````python
def fn(a,b):
    print('a=', a)
    print('b=', b)
    return a+b
````

````python
myr.reduce(fn)
````

````python
myr.reduce(lambda a,b: a+b)
````
