````python
from pyspark import SparkContext
````

````python
sc= SparkContext('local')  
````

````python
my= [10,20,30,40,50]
nRdd= sc.parallelize(my)
nRdd
````

## Actions
- 참조 변수로 반환되지 않는다.
- 따라서 collect() 사용을 안 해도 된다.
- 새로운 객체로 저장이 가능하다.

#### .take( )
- 앞에서 몇개를 가져올 건지

````python
nRdd.take(2)
````

#### .foreach(funtion)
- RDD 안 값들마다 함수를 적용시켜서 반환
- cmd에 값 출력

````python
nRdd.foreach(lambda v: print(v))
````

## Transformations
- 참조 변수(RDD)로 반환된다.
- 따라서 collect()를 사용해야 참조 변수의 저장된 값을 확인할 수 있다.
- 하지만 collect()한 것을 객체로 받으면 안 된다.
- Actions 함수보다 리소스를 덜 잡아 먹는다.
```
map, flatMap, filter...
```

#### .map(function) 
- Actions의 foreach와 유사

````python
newRdd= nRdd.map(lambda v: v+1)
newRdd
````

````python
newRdd.collect()
````

- 예제<br>
30보다 크면 1을 더하고 아니면 2를 더한 RDD를 생성하시오.

````python
ansRdd= nRdd.map(lambda v: v+1 if v> 30 else v+2)
````

````python
ansRdd.collect()
````

````python
txtRdd= sc.textFile('data/aa.txt')
````

````python
txtRdd.collect()
````

- RDD를 transformations 함수를 통해 새로운 RDD를 만들 수 있다.

````python
txtRdd1= txtRdd.map(lambda v: v.split())
txtRdd1.collect()
````

#### .flatMap(function)
- 값을 대괄호를 없애고 function에 의해서 반환

````python
txtRdd2= txtRdd.flatMap(lambda v: v.split())
txtRdd2.collect()
````

## Actions

#### .count()
- RDD의 데이터 개수

````python
txtRdd2.count()
````
