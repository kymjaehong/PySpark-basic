### RDD(Resilient Distributed Dataset)
- 데이터를 여러 서버에 나누어서 저장한다.
- 병렬 처리가 가능하다.
- 단위 데이터 처리 과정에서 일부 서버 혹은 데이터에 문제가 발생해도 스스로 에러를 복구할 수 있다.
<br><br>
https://brocess.tistory.com/182

#### RDD 생성 방법
- sc.textFile( )
- sc.parellelize( )

#### 텍스트 파일 -> RDD
- 하둡 내에서 빅데이터를 읽어온다.

````python
# 'hdfs://localhost:9000/data/aa.txt' 

rdd1= sc.textFile('data/aa.txt')
rdd1
````

#### 리스트 -> RDD

````python
my= [10,20,30,40,50]
rdd2= sc.parallelize(my)
rdd2
````

### RDD Operation
- Actions
- Transformations

#### Actions
```count, collect, reduce, save```
- 함부로 사용하면, 메모리가 죽는다.
- 그래서 데이터를 어느정도 필터링 한 후에 사용해야 한다.
- 분산된 데이터에서  리스트로 모두 가져오기 때문이다.
- 참조 변수(RDD)로 나타나지 않는다.

````python
rdd1.collect()
````

````python
rdd2.collect()
````
