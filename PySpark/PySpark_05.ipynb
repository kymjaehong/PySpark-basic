```python
from pyspark import SparkContext
```

```python
sc= SparkContext('local')
```

```python
empList= ['1,kim,000-000-000','2,lee,111-111-111',
          '3,park,222-222-222','4,song,333-333-333',
          '5,han,555-555-555','6,yoon,666-666-666']
          
deptList= ['1,sale','2,developer','3,manager',
           '4,sale1','5,developer1','6,manager1']
```

```python
empRdd= sc.parallelize(empList)
depRdd= sc.parallelize(deptList)
```

```python
empRdd.collect()
```

```python
depRdd.collect()
```

## Transformations

```python
empRdd1= empRdd.map(lambda v: v.split(','))
empRdd1.collect()
```

```python
empRdd2= empRdd1.map(lambda v: (v[0], v[1]+' '+v[2]))
empRdd2.collect()
```

#### 리스트-> 튜플

```python
depRdd1= depRdd.map(lambda v: v.split(',')).map(lambda v: (v[0],v[1])) 
depRdd1.collect()
```

#### .join( )
- 첫 번째 데이터(key)를 기준으로 데이터를 합친다.

```python
joinData= empRdd2.join(depRdd1)
joinData.collect()
```
