```python
from pyspark import SparkContext
```

```python
sc= SparkContext('local')
```

```python
dt= [(1,2),(3,4),(5,6),(1,7),(3,4)]
nRdd= sc.parallelize(dt)
nRdd.collect()
```

## Transformations

#### .reduceByKey(func,func)
- 튜플 내, 0번(key)이 동일한 것끼리의 1번(value) 합

```python
newRdd= nRdd.reduceByKey(lambda a,b: a+b)
newRdd.collect()
```

```python
newRdd1= nRdd.flatMap(lambda v: v)
newRdd1.collect()
```

```python
newRdd2= nRdd.flatMap(lambda v: v).map(lambda v: (v,1))
newRdd2.collect()
```

## Actions

#### .countByKey( )
- Key의 개수

```python
newRdd2.countByKey()
```

## Transformations

#### .sortByKey( )
- key 순으로 정렬
- ascending-> 내림차순, 오름차순 

```python
newRdd2.sortByKey().collect()
```

#### .sortBy(func)
- 정렬 기준을 설정할 수 있다.
- 예) v[0]= key, v[1]= value로 기준 설정

```python
newRdd.sortBy(lambda v: v[1]).collect()
```

각 글자의 개수를 구하시오.

```python
txtRdd = sc.textFile('data/aa.txt')
txtRdd.collect()
```

```python
txtRdd2 = txtRdd.flatMap(lambda v: v.split()).map(lambda v: (v,1))
txtRdd2.collect()
```

```python
txtRdd2.countByKey()
```

## transformations 함수 + 함수... + Actions 함수
- 마지막이 Actions 함수라면 collect()를 안 해도 된다.

```python
txtRdd.flatMap(lambda v: v.split()).map(lambda v: (v,1)).countByKey()
```
