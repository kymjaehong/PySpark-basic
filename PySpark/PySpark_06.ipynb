````python
from pyspark import SparkContext
````

````python
sc= SparkContext('local')
````

````python
weightRdd= sc.textFile('data/weight.csv')
weightRdd.collect()
````

## Transformations

#### .filter(func)
- 조건에 해당하는 것만 반환

````python
wRdd= weightRdd.map(lambda v: v.split(',')).filter(lambda v: v[0]== '2017')
wRdd.collect()
````

1.전체 키의 합과 평균을 구하시오<br>
2.다음과 같이 변경하시오(학년별 갯수)
- [('elementary', 2), ('middle', 2), ('high', 2)]
3.몸무게가 높은 순으로 정렬하시오

- 1.

````python
tallRdd= weightRdd.map(lambda v: v.split(','))
tallRdd= tallRdd.collect()[1:]
tallRdd
````

````python
sum=0
for i in tallRdd:
    sum+= float(i[1])
````

````python
mean= sum/len(tallRdd)
sum, mean
````

````python
wRdd.map(lambda x: x[1]).map(lambda x: float(x)).sum()
````

````python
wRdd.map(lambda x: x[1]).map(lambda x: float(x)).mean()
````

