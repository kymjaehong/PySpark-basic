```python
from pyspark.sql import  SQLContext, Row
from pyspark import SparkContext,SparkConf
import pandas as pd
import pyspark.sql.functions as f
from pyspark.sql.types import DoubleType,IntegerType,StringType
```

```python
sc= SparkContext('local')
sqlCtx= SQLContext( sc )
```

```python
df1= sqlCtx.read.csv('data/2014년 졸음운전 교통사고.csv', inferSchema= True, header= True,encoding= 'euc-kr')
df2= sqlCtx.read.csv('data/2015년 졸음운전 교통사고.csv', inferSchema= True, header= True,encoding= 'euc-kr')
df3= sqlCtx.read.csv('data/2016년 졸음운전 교통사고.csv', inferSchema= True, header= True,encoding= 'euc-kr')
```

```python
df= df1.unionAll(df2)
df= df.unionAll(df3)
df.show(df.count())
```

## timestamp 변경
- 2016/12/13  : y/M/d
- 2016_12_13 12:12:13  : y_M_d H:m:s

https://docs.oracle.com/javase/tutorial/i18n/format/simpleDateFormat.html

#### f.to_timestamp('컬럼', '컬럼 내 시간 데이터 형식')

```python
df= df.withColumn('tmparse', f.to_timestamp('구분', "yyyy년M월"))
df.show()
```

```python
df.printSchema()
```

```python
df.where(df['tmparse']>= '2015').show()
```

```python
df.where(df['tmparse']>= '2015-05').show()
```

### f.dayofweek(f.col('컬럼'))
- 1:일 2:월 3:화 4:수 5:목 6:금 7:토

#### f.col('컬럼')
- dataframe에서 특정 컬럼을 가져올 때

```python
df.withColumn('f1', f.date_format(df['tmparse'], 'yyyy MM dd EEEE')).\
    withColumn('f2', f.dayofweek(f.col('tmparse'))).show()
```

```python
df= df.withColumn('w', f.dayofweek(f.col('tmparse')))
df.show()
```

```python
w= {1:'일',2:'월',3:'화',4:'수',5:'목',6:'금',7:'토'}
ff= f.udf(lambda v: w[v])
df.withColumn('요일', ff(df['w'])).show()
```

```python
df.withColumn('year', f.date_format(df['tmparse'], 'yyyy'))\
.withColumn('month', f.date_format(df['tmparse'], 'MM'))\
.groupBy(['year', 'month']).avg('사망(명)').orderBy(['year','month']).show()
```
